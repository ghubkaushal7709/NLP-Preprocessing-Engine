# Task 1: Conceptual Understanding

## 1. Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" are treated as different words because of case sensitivity.
To avoid duplication and improve consistency, text is usually converted to lowercase.

## 2. What happens if stopwords are not removed?
If stopwords are not removed, the dataset becomes noisy and large.
Common words like "is", "the", "and" do not add much meaning and may reduce model performance.

## 3. When removing stopwords is harmful
- Sentiment Analysis: "I am not happy" → removing "not" changes meaning
- Question Answering: "What is your name?" → removing "what" breaks intent

## 4. Stemming vs Lemmatization
Stemming cuts words to root form (not always meaningful), while lemmatization converts words to proper dictionary form using context.

```python
# Task 2: Build Advanced Preprocessing Function

import re

def preprocess_text(text):
    """
    This function cleans and preprocesses raw text data.
    It handles noise like numbers, URLs, repeated characters, etc.
    """

    # Error Handling: Check if input is valid
    if not text or not isinstance(text, str):
        return [], ""

    # 1. Remove URLs and email patterns
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    # 2. Remove numbers
    text = re.sub(r"\d+", "", text)

    # 3. Convert to lowercase
    text = text.lower()

    # 4. Handle repeated characters (e.g., "soooo" → "so")
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 5. Remove special characters and emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # 6. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 7. Tokenization (split into words)
    tokens = text.split()

    # 8. Remove short tokens (length ≤ 2 except "no", "not")
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # 9. Recreate cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
# Task 3: Testing preprocessing on diverse sentences

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Apply preprocessing and display results
for text in test_sentences:
    tokens, clean = preprocess_text(text)

    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 60)
```

```python
# Task 4: Compute token statistics

import numpy as np

analytics = []

for text in test_sentences:
    tokens, clean = preprocess_text(text)

    # Calculate metrics
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens]) if tokens else 0

    # Store results
    analytics.append({
        "Original Text": text,
        "Total Tokens": total_tokens,
        "Unique Tokens": unique_tokens,
        "Average Token Length": avg_length
    })

# Display results
for item in analytics:
    print(item)
```

```python
# Task 5: Word frequency analysis

from collections import Counter

all_tokens = []

# Combine tokens from all sentences
for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

# Count frequency
counter = Counter(all_tokens)

# Top 10 frequent words
print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

# Top 5 least frequent words
print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
# Task 6: Build full pipeline

def full_pipeline(text_list):
    """
    This function processes a list of sentences
    and returns all tokens and cleaned sentences.
    """

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Test pipeline
result = full_pipeline(test_sentences)

print("All Tokens:", result["tokens"])
print("\nClean Sentences:", result["clean_sentences"])
```

```python
# Task 7: Test edge cases

edge_cases = [
    "",             # Empty string
    "😂😂😂😂",       # Only emojis
    "123456789",    # Only numbers
]

for case in edge_cases:
    tokens, clean = preprocess_text(case)

    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 50)
```
