```python
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration
```

```python
model_name = "facebook/blenderbot-400M-distill"

tokenizer = BlenderbotTokenizer.from_pretrained(model_name)
model = BlenderbotForConditionalGeneration.from_pretrained(model_name)
```

```python
print("Chatbot: Hello! I am your AI assistant. How can I help you today?\n")

conversation_history = ""

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! Have a great day 😊")
        break

    # Add user input to conversation history
    conversation_history += "User: " + user_input + " </s> "

    # Tokenize input
    inputs = tokenizer(conversation_history, return_tensors="pt", truncation=True)

    # Generate response
    reply_ids = model.generate(
        **inputs,
        max_length=200,
        do_sample=True,
        top_k=50,
        top_p=0.9,
        temperature=0.7
    )

    # Decode response
    response = tokenizer.decode(reply_ids[0], skip_special_tokens=True)

    print("Chatbot:", response)

    # Add response to history
    conversation_history += "Bot: " + response + " </s> "
```
